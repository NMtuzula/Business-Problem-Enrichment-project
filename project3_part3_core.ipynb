```python
import pandas as pd
from sqlalchemy import create_engine

from sqlalchemy_utils import database_exists, create_database
import pymysql
pymysql.install_as_MySQLdb()
from urllib.parse import quote_plus as urlquote
```

```python
connection = "mysql+pymysql://root:root@localhost/movies"
engine = create_engine(connection)
```

```python
database_exists(connection)
```

```python
title_basics = pd.read_csv('Data/title_basic.csv.gz')
title_basics.head()
```

```python
title_ratings = pd.read_csv('Data/title_ratings.csv.gz')
title_ratings.head()
```

```python
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
tmdb_data = pd.concat([df_2000, df_2001])
tmdb_data.head()
```

```python
## create a col with a list of genres
title_basics['genres_split'] = title_basics['genres'].str.split(',')
title_basics
```

```python
exploded_genres = title_basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
```

```python
## Save just tconst and genres_id as new df
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
title_basics['genre_id'] = title_basics['genres_split'].replace(genre_map)
title_basics = title_basics.drop(columns='genres_split')
```

```python
## Manually make dataframe with named cols from the .keyd and .values
genres = pd.DataFrame(genre_map.items(), columns=['Genre_Name', 'Genre_ID'])
genres.head()
```

Discard unnecessary information:

```python
title_basics = title_basics.drop(columns=['originalTitle', 'isAdult', 'titleType', 'genres', 'endYear'])
title_basics
```

```python
# will convert genre_id to string
title_basics['genre_id'] = title_basics['genre_id'].astype(str)
```

```python
title_basics['genre_id'] = title_basics['genre_id'].str.replace("[",'')
title_basics['genre_id'].head()
```

```python
title_basics['genre_id'] = title_basics['genre_id'].str.replace("]",'')
title_basics['genre_id'].head()
```

```python
title_basics
```

```python
# keep certain columns in TMDB_API_Results
tmdb_data = tmdb_data[['imdb_id', 'budget', 'revenue', 'certification']]                                        
tmdb_data
```

```python
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = title_basics['tconst'].fillna('').map(len).max()
title_len = title_basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_basics = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    "genre_id":Text(title_len+1),
    'startYear':Float(),
    'runtimeMinutes':Integer()}

## Calculate max string lengths for object columns
key_len1 = title_ratings['tconst'].fillna('').map(len).max()
df_ratings = {
    "tconst": String(key_len1+1),
    'averageRating' :Float(),
    'numVotes' :Integer()}

## Calculate max string lengths for object columns
title_len1 = tmdb_data['imdb_id'].fillna('').map(len).max()
df_tmdb = {
    'imdb_id': Text(title_len1+1),
    'budget': Float(),
    'revenue': Float(),
    'certification': Text(title_len+1)}

title_len3 = genres['Genre_Name'].fillna('').map(len).max()
df_genres = {
    'Genre_ID': Integer(),
    'Genre_Name': Text(title_len3+1)}
```

Save to sql and index=False

```python
title_basics.to_sql('title_basics',engine,dtype=df_basics,if_exists='replace',index=False)
```

```python
title_ratings.to_sql('title_ratings',engine,dtype=df_ratings,if_exists='replace',index=False)
```

```python
tmdb_data.to_sql('tmdb_data',engine,dtype=df_tmdb,if_exists='replace',index=False)
```

```python
genres.to_sql('genres',engine,dtype=df_genres,if_exists='replace',index=False)
```

```python
title_genres.to_sql('title_genres',engine,if_exists='replace',index=False)
```
